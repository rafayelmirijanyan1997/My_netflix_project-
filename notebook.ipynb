## The data
### **netflix_data.csv**

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like


print(netflix_df['type'] == 'Movie')
```

```python
netflix_df_movies = netflix_df[(netflix_df['type'] == 'Movie')]
netflix_df_movies
```

So 4677 out of 4812 are actually movies, let's try to find out movies where the release year is between 1990 and 1999

```python
import numpy as np

filtered_df = np.logical_and(
    (netflix_df_movies['release_year'] >= 1990), 
    (netflix_df_movies['release_year'] < 2000)

)
netflix_df_1990s = netflix_df_movies[filtered_df]
netflix_df_1990s
```

Only 183 Movies were released in 1990s. So now lets visualize the data to understand the distribution of hte durations.

```python
import matplotlib.pyplot as plt
import numpy as np


# Create a figure and axis for customization
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the histogram
n, bins, patches = ax.hist(netflix_df_1990s['duration'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)

# Add a title and labels
ax.set_title('Distribution of Movie Durations in 1990s', fontsize=16, fontweight='bold')
ax.set_xlabel('Duration (minutes)', fontsize=14)
ax.set_ylabel('Frequency', fontsize=14)

# Add gridlines for better readability
ax.grid(True, linestyle='--', alpha=0.7)

# Add a vertical line for the mean
mean_duration = np.mean(netflix_df_1990s['duration'])
ax.axvline(mean_duration, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_duration:.1f} min')

# Add a vertical line for the median
duration = int(np.median(netflix_df_1990s['duration']))
ax.axvline(duration, color='green', linestyle='dashed', linewidth=2, label=f'Median: {duration:.1f} min')

# Add a legend to explain the lines
ax.legend(loc='upper right', fontsize=12)

# Customize ticks and limits
ax.set_xticks(np.arange(0, max(netflix_df_1990s['duration']) + 20, step=20))
ax.set_xlim(0, max(netflix_df_1990s['duration']) + 10)

# Add annotations for some key statistics
plt.text(mean_duration + 5, max(n) * 0.9, f'Mean: {mean_duration:.1f}', color='red', fontsize=12)
plt.text(duration + 5, max(n) * 0.8, f'Median: {duration:.1f}', color='green', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
action_movies_1990s = netflix_df_1990s[netflix_df_1990s["genre"] == "Action"]
```

```python
short_movie_count = 0  # Initialize counter

# Iterate over each row in the DataFrame
for index, row in action_movies_1990s.iterrows():
    if row['duration'] < 90:
        short_movie_count += 1 
    else:
        short_movie_count = short_movie_count

print(f"Number of movies with duration less than 90 minutes: {short_movie_count}")
```

So only 7 movies that were released in 1990s have duration less than 90 minutes, saved in short_movie_count
